Napisz program znajdujący _kwadrat magiczny_ o wymiarach 4 x 4, w którym każda
komórka zawiera liczby z zakresu 1..9 (mogą się powtarzać). Dodatkowo,
suma liczb dla każdego wiersza i kolumny wynosi 20.


```python
% Definiujemy zakres liczb, które mogą być używane (1-9).
number(1..9).

% Tworzymy kwadrat 4x4, gdzie każda komórka ma przypisaną wartość.
% cell(X, Y, V) oznacza, że w komórce (X, Y) znajduje się wartość V.
{ cell(X, Y, V) : number(V) } = 1 :- X=1..4, Y=1..4.

% Warunek, że suma wartości w każdym wierszu musi wynosić 20.
:- 20 != #sum { V, Y : cell(X, Y, V) }, X=1..4.

% Warunek, że suma wartości w każdej kolumnie musi wynosić 20.
:- 20 != #sum { V, X : cell(X, Y, V) }, Y=1..4.

% Rozwiązujemy problem.
#show cell/3.
```

    1.Definiuje możliwe wartości do umieszczenia w kwadracie (od 1 do 9).

    3.Zapewnia, że suma wartości w każdym wierszu i każdej kolumnie wynosi dokładnie 20.

    4.Wyświetla rozwiązanie.

Mój program CLINGO pomyślnie znalazł rozwiązanie, które spełnia zadane kryteria: każda komórka zawiera liczbę z zakresu 1-9, a suma liczb w każdym wierszu i kolumnie wynosi 20. Jest to jeden z możliwych kwadratów magicznych, który spełnia te warunki.


Zmodyfikowałem też poprzedni kod aby znalazł inny taki kwadrat.
